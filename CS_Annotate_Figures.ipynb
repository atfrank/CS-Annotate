```
from sklearn.metrics import accuracy_score,f1_score,balanced_accuracy_score,recall_score,precision_score,classification_report,confusion_matrix
def calculate_score_chain_classifier(target,cv):
  # load in predictions
  #targets = ['astack', 'nastack', 'pair', "C1' exo", "C2' endo", "C2' exo", "C3' endo", "C3' exo", "C4' exo", 'sasa', 'syn/anti']
  
  y_test = pd.read_csv(DIR_PATH+"chain/y_test_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_1 = pd.read_csv(DIR_PATH+"chain/pred_chain_0_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_2 = pd.read_csv(DIR_PATH+"chain/pred_chain_1_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_3 = pd.read_csv(DIR_PATH+"chain/pred_chain_2_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_4 = pd.read_csv(DIR_PATH+"chain/pred_chain_3_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_5 = pd.read_csv(DIR_PATH+"chain/pred_chain_4_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_6 = pd.read_csv(DIR_PATH+"chain/pred_chain_5_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_7 = pd.read_csv(DIR_PATH+"chain/pred_chain_6_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_8 = pd.read_csv(DIR_PATH+"chain/pred_chain_7_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_9 = pd.read_csv(DIR_PATH+"chain/pred_chain_8_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  chain_10 = pd.read_csv(DIR_PATH+"chain/pred_chain_9_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  ensemble = pd.read_csv(DIR_PATH+"chain/pred_ensemble_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  ovr = pd.read_csv(DIR_PATH+"chain/pred_ovr_cv_"+str(cv)+".csv",delim_whitespace=True,header=None)#,names=targets)
  # calculate balanced accuracy
  score_ovr = balanced_accuracy_score(y_test[target], ovr[target]>=0.5)
  score_chain_1 = balanced_accuracy_score(y_test[target], chain_1[target]>=0.5)
  score_chain_2 = balanced_accuracy_score(y_test[target], chain_2[target]>=0.5)
  score_chain_3 = balanced_accuracy_score(y_test[target], chain_3[target]>=0.5)
  score_chain_4 = balanced_accuracy_score(y_test[target], chain_4[target]>=0.5)
  score_chain_5 = balanced_accuracy_score(y_test[target], chain_5[target]>=0.5)
  score_chain_6 = balanced_accuracy_score(y_test[target], chain_6[target]>=0.5)
  score_chain_7 = balanced_accuracy_score(y_test[target], chain_7[target]>=0.5)
  score_chain_8 = balanced_accuracy_score(y_test[target], chain_8[target]>=0.5)
  score_chain_9 = balanced_accuracy_score(y_test[target], chain_9[target]>=0.5)
  score_chain_10 = balanced_accuracy_score(y_test[target], chain_10[target]>=0.5)
  ensemble_score = balanced_accuracy_score(y_test[target], ensemble[target]>=0.5)
  return [score_ovr,score_chain_1,score_chain_2,score_chain_3,score_chain_4,score_chain_5,score_chain_6,score_chain_7,score_chain_8,score_chain_9,score_chain_10,ensemble_score]

def draw_ensemble(target, scores): 
  models=["Validation Set 1", "Validation Set 2","Validation Set 3","Validation Set 4","Validation Set 5"]
  colors = ['indianred','lightsalmon','lightsalmon','lightsalmon','lightsalmon','lightsalmon','lightblue']
  fig = go.Figure()
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Base Model"]],
      name='Base Model',
      marker_color='indianred')
  )
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 1"]],
      name='Chain 1',
      marker_color='lightsalmon')
  )
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 2"]],
      name='Chain 2',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 3"]],
      name='Chain 3',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 4"]],
      name='Chain 4',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 5"]],
      name='Chain 5',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 6"]],
      name='Chain 6',
      marker_color='lightsalmon')
  )
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 7"]],
      name='Chain 7',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 8"]],
      name='Chain 8',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 9"]],
      name='Chain 9',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Chain 10"]],
      name='Chain 10',
      marker_color='lightsalmon'
  ))
  fig.add_trace(go.Bar(
      x=models,
      y=[round(num,3) for num in scores["Ensemble"]],
      name='Ensemble',
      marker_color='lightblue'
  ))

  fig.update_yaxes(range=[scores.min().min()*0.9, scores.max().max()*1.1])
  fig.update_layout(template="plotly_white",font=dict(family="Helvetica",size=36),
                    legend=dict(
                        x=1,
                        y=1,
                        traceorder="normal",
                        bgcolor="LightGrey",
                        bordercolor="Black",
                        borderwidth=2
                    ),
                    title={
                        'text': target,
                        'y':0.8,
                        'x':0.5,
                        'xanchor': 'center',
                        'yanchor': 'top'},
                    width=2000,
                    height=800
                    )
  fig.update_yaxes(showgrid=False)
  fig.show()

def get_cv_scores(i, verbose=True):
  scores_1 = calculate_score_chain_classifier(i, 1)
  scores_2 = calculate_score_chain_classifier(i, 2)
  scores_3 = calculate_score_chain_classifier(i, 3)
  scores_4 = calculate_score_chain_classifier(i, 4)
  scores_5 = calculate_score_chain_classifier(i, 5)
  targets = ['Adjacent Stacking', 'Non-adjacent Stacking', 'Base Pairing', "C1'-exo", "C2'-endo",
        "C2'-exo", "C3'-endo", "C3'-exo", "C4'-exo",
        'SASA', 'Syn or Anti']
  scores = pd.DataFrame([scores_1,scores_2,scores_3,scores_4,scores_5])
  scores.columns = ["Base Model","Chain 1","Chain 2","Chain 3","Chain 4","Chain 5","Chain 6","Chain 7","Chain 8","Chain 9","Chain 10","Ensemble"]
  scores.index = ["Validation 1","Validation 2","Validation 3","Validation 4","Validation 5"]
  
  # this line for plotting validation score for each chain + ensemble classifier
  if verbose:
    draw_ensemble(target = targets[i], scores = scores) 
  # this line for overlaying chainclassifier and deepchem model
  return(scores) 
```

```
# draw performance of classifier chain
for i in range(11):
  get_cv_scores(i, verbose=True)
```

```
# compare ensemble and base learner
targets = ['astack', 'nastack', 'pair', "C1'-exo", "C2'-endo",
       "C2'-exo", "C3'-endo", "C3'-exo", "C4'-exo",
       'sasa', 'syn/anti']
scores_avg = []
for i in range(11):
  scores = get_cv_scores(i, verbose=False) # set verbose to False, do not draw figures
  avg = scores.apply(np.mean, axis=0).tolist()
  scores_avg.append(avg)

scores_avg = pd.DataFrame(scores_avg)
scores_avg.columns = ["Base Model","Chain 1","Chain 2","Chain 3","Chain 4","Chain 5","Chain 6","Chain 7","Chain 8","Chain 9","Chain 10","Ensemble"]
scores_avg.index = targets

fig = go.Figure()
fig.add_trace(go.Bar(y=scores_avg["Base Model"],x=targets,name="Independent Classifiers",marker_color='rgb(128,177,211)'))#,text=text1,mode="lines+markers+text",textposition="top center"))
fig.add_trace(go.Bar(y=scores_avg["Ensemble"],x=targets,name="Chained Ensemble",marker_color='rgb(253,180,105)'))
fig.update_layout(
    template = 'plotly_white',
    font=dict(family="Helvetica",size=28),width=1400,height=600,
    legend=dict(
        x=0.8,
        y=1.3,
        traceorder="normal",
        bgcolor="LightGrey",
        bordercolor="Black",
        borderwidth=2
        )
    )
fig.show()
```

```
rna="2N1Q"
actual = pd.read_csv(DIR_PATH+"actual_"+rna+".csv",header=0,delim_whitespace=True)
pred = pd.read_csv(DIR_PATH+"pred_"+rna+".csv",header=0,delim_whitespace=True)
```

```
for rna in ['2JTP','2LU0','5KH8','2N1Q']:
  actual = pd.read_csv(DIR_PATH+"actual_"+rna+".csv",header=0,delim_whitespace=True)
  pred = pd.read_csv(DIR_PATH+"pred_"+rna+".csv",header=0,delim_whitespace=True)
  # False negative
  FN = (actual == True) & (pred == False)
  # False positive
  FP = (actual == False) & (pred == True)
  # True positive/negative
  TP_and_TN = (actual == True) & (pred == True) | (actual == False) & (pred == False)

  result = FN.copy()
  result.where(FN.values==False, 0, inplace=True)
  result.where(FP.values==False, 0.5, inplace=True)
  result.where(TP_and_TN.values==False, 1, inplace=True)
  print(result.head())

  fig = go.Figure(
      data=go.Heatmap(
      z = result,
      x=["sasa","astack","nstack","pair","syn/anti","C2'-endo","C3'-endo","C2'-exo","C3'-exo","C1'-exo","C4'-exo"],
      colorscale = 'deep',
      showscale = False
      ))
  
  fig.update_layout(
      yaxis = dict(title_text = "Residues"),
      xaxis = dict(title_text = "Features"),
      title={'text':rna,'y':0.96,'x':0.5},
      #width = 2100,
      #height=1200,
      font = dict(family='Helvetica',size=24),
      xaxis_nticks=24)
  fig.show()
```

```
predictions = pd.read_csv(DIR_PATH+"deepchem_testset_predictions_update_2N1Q.txt",delim_whitespace=True,header=0)
print(predictions.columns)
actual = predictions[['astack', 'nastack', 'pair', 'pucker_C1p_exo', 'pucker_C2p_endo', 'pucker_C2p_exo', 'pucker_C3p_endo', 'pucker_C3p_exo', 'pucker_C4p_exo','sasa.1','syn_anti']]
pred = predictions[['pastack', 'pnastack', 'ppair', 'ppucker_C1p_exo', 'ppucker_C2p_endo', 'ppucker_C2p_exo', 'ppucker_C3p_endo', 'ppucker_C3p_exo', 'ppucker_C4p_exo','psasa','psyn_anti']]
```

```
targets = ['astack','nastack','pair','pucker_C1p_exo','pucker_C2p_endo','pucker_C2p_exo', 'pucker_C3p_endo', 'pucker_C3p_exo','pucker_C4p_exo','sasa','syn_anti']
pred.columns = targets
actual.columns = targets
```

```
recall1=[]
recall0=[]
balanced_accuracys=[]

for i,target in enumerate(targets):
  recall1.append(recall_score(actual[target], pred[target]>=0.5))
  recall0.append(recall_score(actual[target], pred[target]>=0.5,pos_label=0))
  balanced_accuracys.append(balanced_accuracy_score(actual[target], pred[target]>=0.5)) 
test_stats = pd.DataFrame({'sensitivity': recall1, 'specificity':recall0, 'balanced accuracy':balanced_accuracys})
test_stats.index = targets
test_stats.round(3)
```

```
acc = []
for rna in ['2JTP','2LU0','2N1Q','5KH8']:
  balanced_accuracys=[]
  actual = pd.read_csv(DIR_PATH+'actual_'+rna+'.csv',delim_whitespace=True,header=0)
  pred = pd.read_csv(DIR_PATH+'pred_'+rna+'.csv',delim_whitespace=True,header=0)
  for i,target in enumerate(targets):
    balanced_accuracys.append(balanced_accuracy_score(actual[target], pred[target]>=0.5)) 
  acc.append(balanced_accuracys)
```
